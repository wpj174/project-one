```python
import pandas as pd
```

```python
GB_youtube_csv = "Resources/GB_Merged_data.csv"
```

```python
GB_youtube_DF = pd.read_csv(GB_youtube_csv, low_memory=False)
GB_youtube_DF.head()
```

```python
#https://www.kite.com/python/answers/how-to-calculate-a-pandas-dataframe-time-difference-between-two-columns-in-python
GB_youtube_DF_timeDiff = pd.DataFrame(columns=["publishedAt", "trending_date"])

GB_youtube_DF_timeDiff.publishedAt = GB_youtube_DF.loc[:,"publishedAt"]
GB_youtube_DF_timeDiff.publishedAt = pd.to_datetime(GB_youtube_DF_timeDiff.publishedAt)

GB_youtube_DF_timeDiff.trending_date = GB_youtube_DF.loc[:, "trending_date"]
GB_youtube_DF_timeDiff.trending_date = pd.to_datetime(GB_youtube_DF_timeDiff.trending_date)

difference = pd.DataFrame(GB_youtube_DF_timeDiff.trending_date - GB_youtube_DF_timeDiff.publishedAt)
difference.head()
```

```python
GB_youtube_DF['Time to Trend'] = difference[0].values
GB_youtube_DF.head()
```
