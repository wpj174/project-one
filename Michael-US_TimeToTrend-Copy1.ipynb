```python
import pandas as pd
```

```python
US_youtube_csv = "Resources/US_Merged_data.csv"
```

```python
US_youtube_DF = pd.read_csv(US_youtube_csv, low_memory=False)
US_youtube_DF.head()
```

```python
#https://www.kite.com/python/answers/how-to-calculate-a-pandas-dataframe-time-difference-between-two-columns-in-python
US_youtube_DF_timeDiff = pd.DataFrame(columns=["publishedAt", "trending_date"])

US_youtube_DF_timeDiff.publishedAt = US_youtube_DF.loc[:,"publishedAt"]
US_youtube_DF_timeDiff.publishedAt = pd.to_datetime(US_youtube_DF_timeDiff.publishedAt)

US_youtube_DF_timeDiff.trending_date = US_youtube_DF.loc[:, "trending_date"]
US_youtube_DF_timeDiff.trending_date = pd.to_datetime(US_youtube_DF_timeDiff.trending_date)

difference = pd.DataFrame(US_youtube_DF_timeDiff.trending_date - US_youtube_DF_timeDiff.publishedAt)
difference
```

```python
US_youtube_DF['Time to Trend'] = difference[0].values
US_youtube_DF.head()
```
