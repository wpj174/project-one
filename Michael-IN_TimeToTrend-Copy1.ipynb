```python
import pandas as pd
```

```python
IN_youtube_csv = "Resources/IN_Merged_data.csv"
```

```python
IN_youtube_DF = pd.read_csv(IN_youtube_csv, low_memory=False)
IN_youtube_DF.head()
```

```python
#https://www.kite.com/python/answers/how-to-calculate-a-pandas-dataframe-time-difference-between-two-columns-in-python
IN_youtube_DF_timeDiff = pd.DataFrame(columns=["publishedAt", "trending_date"])

IN_youtube_DF_timeDiff.publishedAt = IN_youtube_DF.loc[:,"publishedAt"]
IN_youtube_DF_timeDiff.publishedAt = pd.to_datetime(IN_youtube_DF_timeDiff.publishedAt)

IN_youtube_DF_timeDiff.trending_date = IN_youtube_DF.loc[:, "trending_date"]
IN_youtube_DF_timeDiff.trending_date = pd.to_datetime(IN_youtube_DF_timeDiff.trending_date)

difference = pd.DataFrame(IN_youtube_DF_timeDiff.trending_date - IN_youtube_DF_timeDiff.publishedAt)
difference.head()
```

```python
IN_youtube_DF['Time to Trend'] = difference[0].values
IN_youtube_DF.head()
```
